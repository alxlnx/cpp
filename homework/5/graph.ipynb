# График зависимости времени сложения 10 000 000 чисел от количества использованных ядер

```python
# amount_of_nums    threads_num    time (nanoseconds)

import fnmatch
import os
import matplotlib.pyplot as plt
import numpy as np

font = {'fontname': 'DejaVu Serif'}
FILE_AMOUNT  = 100

threads_O0   = []
times_O0     = []
threads_O1   = []
times_O1     = []
threads_O2   = []
times_O2     = []
threads_O3   = []
times_O3     = []
#########################################################
for i in range(FILE_AMOUNT):
    filename = 'homework_5_data_' + str(i + 1) + '_O0.txt'
    with open(filename, 'r') as f:
        for line in f:
            data = line.rstrip('\n').split('\t')
            
            threads_O0.append(data[1])
            times_O0.append(data[2])

threads_O0 = np.array(threads_O0)
threads_O0 = threads_O0.astype('int32')
times_O0   = np.array(times_O0)
times_O0   = times_O0.astype('int32')

#########################################################
for i in range(FILE_AMOUNT):
    filename = 'homework_5_data_' + str(i + 1) + '_O1.txt'
    with open(filename, 'r') as f:
        for line in f:
            data = line.rstrip('\n').split('\t')
            
            threads_O1.append(data[1])
            times_O1.append(data[2])

threads_O1 = np.array(threads_O1)
threads_O1 = threads_O1.astype('int32')
times_O1   = np.array(times_O1)
times_O1   = times_O1.astype('int32')

#########################################################
for i in range(FILE_AMOUNT):
    filename = 'homework_5_data_' + str(i + 1) + '_O2.txt'
    with open(filename, 'r') as f:
        for line in f:
            data = line.rstrip('\n').split('\t')            
            
            threads_O2.append(data[1])
            times_O2.append(data[2])

threads_O2 = np.array(threads_O2)
threads_O2 = threads_O2.astype('int32')
times_O2   = np.array(times_O2)
times_O2   = times_O2.astype('int32')

#########################################################
for i in range(FILE_AMOUNT):
    filename = 'homework_5_data_' + str(i + 1) + '_O3.txt'
    with open(filename, 'r') as f:
        for line in f:
            data = line.rstrip('\n').split('\t')
            
            threads_O3.append(data[1])
            times_O3.append(data[2])

threads_O3 = np.array(threads_O3)
threads_O3 = threads_O3.astype('int32')
times_O3   = np.array(times_O3)
times_O3   = times_O3.astype('int32')

fig, ax = plt.subplots(figsize=(10,7), dpi=400)
ax.grid(which='major', linestyle='--', linewidth=0.5)
ax.minorticks_on()
ax.set_xlabel('Число потоков, шт', **font)
ax.set_ylabel('Время суммирования, мкс', **font)
ax.set_title('Рис 1. График для уровней оптимизации O0, O1, O2, O3', **font)
ax.scatter(threads_O0, times_O0, c='k', zorder=3, label='O0')
ax.scatter(threads_O1, times_O1, c='b', zorder=3, label='O1')
ax.scatter(threads_O2, times_O2, c='r', zorder=3, label='O2')
ax.scatter(threads_O3, times_O3, c='g', zorder=3, label='O3')
ax.legend()
```

```python
threads_O3_NANO   = []
times_O3_NANO     = []
#########################################################
for i in range(FILE_AMOUNT):
    filename = 'homework_5_data_' + str(i + 1) + '_O3_NANO.txt'
    with open(filename, 'r') as f:
        for line in f:
            data = line.rstrip('\n').split('\t')
            
            threads_O3_NANO.append(data[1])
            times_O3_NANO.append(data[2])

threads_O3_NANO = np.array(threads_O3_NANO)
threads_O3_NANO = threads_O3_NANO.astype('int32')
times_O3_NANO   = np.array(times_O3_NANO)
times_O3_NANO   = times_O3_NANO.astype('int32')
fig, ax = plt.subplots(figsize=(10,7), dpi=400)
ax.grid(which='major', linestyle='--', linewidth=0.5)
ax.minorticks_on()
ax.set_title('Рис 2. График для уровня оптимизации O3', **font)
ax.set_xlabel('Число потоков, шт', **font)
ax.set_ylabel('Время суммирования, нс', **font)
ax.scatter(threads_O3_NANO, times_O3_NANO, c='g', zorder=3, label='O3')
ax.legend()
```

```python
one_core_times = []
for i in range(int(len(times_O3_NANO) / 24)):
    one_core_times.append( times_O3_NANO[24 * i] )
avg_time = np.mean(one_core_times)
print(f'Среднее время суммирования на одном потоке при уровне оптимизации O3: { int(avg_time) } нс')
```
